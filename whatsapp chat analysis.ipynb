```python
f = open("D:\PYTHON PRACTICE\PROJECT\DATASET\WhatsApp Chat with Shahzeb AMU\WhatsApp Chat with Shahzeb AMU.txt", 'r', encoding="utf-8")
data = f.read()
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern,data)[1:]
```

```python
datess = re.findall(pattern,data)
datess
```

```python
df = pd.DataFrame({'user_message':messages,"message_date":datess})
```

```python
# seperate users and messages
users =[]
mess = []
count =0
for messages in df["user_message"]:
    entry = re.split('([\w\W]+?):\s',messages)
    if entry[1:]:
        users.append(entry[1])
        mess.append(entry[2])
    else:
        users.append("group notification")
        mess.append(entry[0])
```

```python
df["user"]=users
```

```python
df["message"]=mess
```

```python
df.drop(columns=['user_message'],inplace=True)
```
